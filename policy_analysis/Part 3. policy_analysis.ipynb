### 3.2. Policy  recommendations

 
**1. From Needs Improvement to Good**

Factors that “pull down” performance are low levels of digitalization (sdg9_digital), weak scores in corruption perception (sdg16_cpi), insufficient R&D investment (sdg9_eurd), and lagging higher education outcomes (sdg4_tertiary).

**Programs aimed at economic growth, SME development, and youth employment**
- retraining programs for the unemployed  
- support for small and medium-sized enterprises (SMEs)  
- investment incentives for green industries  
- tax benefits for job creation  
- state internship programs for young people  
- stable microcredit programs for startups and small businesses  

**Programs for infrastructure development and digitalization**
- expanding access to the internet and digital skills  
- programs engaging older people in the active use of digital products  
- national broadband development plans  
- grants for digitalization of schools and local communities  
- creation of regional digital literacy centers  
- subsidies for SME digitalization  
- free online course programs  
- digitalization programs in the healthcare sector  
- partnerships with private companies to provide internet access in remote areas  

**Programs for social justice and inclusion of vulnerable groups**
- scholarship programs for children in need of additional financial support  
- support for gender and inclusive equality: monitoring programs on women’s rights and the rights of people with special needs in the labor market, as well as programs monitoring pay equity   
- programs for the integration and active inclusion of migrants  
- affordable housing programs for low-income groups  
- free access to basic healthcare services  
- family support programs for households with children

**2. From Good to Leaders**

Factors: a significant impact of broadband penetration (sdg9_bband) and CO₂ intensity (sdg13_co2gcp), with a moderate effect from renewables (sdg7_eurenew).

**Innovation and Education Development – Strengthening Competitiveness through R&D and Knowledge**
- national programs and funds for research financing  
- university–business startup incubators  
- mobility programs for students and researchers  
- tax credits for innovative investments  
- regional innovation clusters  
- programs encouraging women to enter “traditionally non-female” professional fields  
- financing of STEM education in schools  

**Energy and Climate Transition – Reducing Emissions and Scaling up Renewables**
- subsidy programs for solar panels for private households as well as small and medium-sized enterprises (SMEs)  
- the 'Green Deal' for public and private transport  
- support for energy-efficient building modernization  
- development of offshore wind energy  
- promotion of low-emission public transport  

**Institutions and Integrity – Strengthening Trust in Government and Transparency**
- anti-corruption agencies with independent mandates  
- open data on public expenditures (e-Data) 
- media programs supporting independent journalism  
- monitoring compliance with legislation on freedom of the press, as well as protection of whistleblowers  
- safeguarding the right to peaceful assembly and protest  
- judicial system reforms  
- electronic governance (e-Gov)  
- development of independent auditing and oversight bodies

**3. Stay Leader**

Factors: positive contributions from R&D (sdg9_eurd), rule of law and press freedom (sdg16_rsf), income per capita (sdg8_income), and adult education (sdg4_adult). Negative influences come from the gender employment gap (sdg5_empgap), crime (sdg16_crime), and CO₂ per GDP (sdg13_co2gcp).

**Global Leadership – Exporting Best Practices in Sustainable Development**
- international technical assistance programs

- development partnerships with countries in need of support  
- joint scientific projects with the Global South  
- financing climate adaptation in vulnerable regions  
- humanitarian education programs  
- global funds for green transformation  

**Social Innovation and Inclusion – Reducing Remaining Gaps in Gender and Income**
- equal pay programs  
- social innovation hubs  
- basic income pilot projects  
- corporate quotas for women in leadership  
- universal child benefits  
- migrant integration through vocational training  
- public campaigns against discrimination  

**Advanced Technologies and Resilience – Scaling Breakthrough Solutions in Digital, Energy, and Health Sectors**
- artificial intelligence for sustainable production  
- smart grids for green energy  
- biotechnology innovations in healthcare  
- IoT-based waste management systems

### 3.3. Country-specific insights

Let's build a dumbbell chart to show the change in SDGI score for each country

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

YEAR_FROM, YEAR_TO = 2015, 2024
TIER01, TIER12 = 65, 75  

df = pd.read_csv("cleaned_data.csv")
df = df[df["year"].isin([YEAR_FROM, YEAR_TO])]

wide = (df.pivot_table(index="Country", columns="year",
                       values="sdgi_score", aggfunc="mean")
          .dropna(subset=[YEAR_FROM, YEAR_TO])
          .sort_values(YEAR_TO))

countries = wide.index.tolist()
y = range(len(countries))

fig, ax = plt.subplots(figsize=(12, len(countries) * 0.3))

def tier(x):
    if x < TIER01: return 0
    elif x < TIER12: return 1
    else: return 2
tier_colors = {0: "red", 1: "orange", 2: "green"}

for i, c in enumerate(countries):
    x0, x1 = wide.loc[c, YEAR_FROM], wide.loc[c, YEAR_TO]
    ax.hlines(y=i, xmin=x0, xmax=x1, color="lightgray")
    ax.scatter(x0, i, color=tier_colors[tier(x0)], edgecolor="black", s=60, zorder=2)
    ax.scatter(x1, i, color=tier_colors[tier(x1)], edgecolor="black", s=60, zorder=2)

for thr in [TIER01, TIER12]:
    ax.axvline(thr, color="gray", linestyle="--", linewidth=1)

xmin, xmax = ax.get_xlim()
ax.text((xmin + TIER01) / 2,   0.975, "Tier 0", ha="center", va="bottom",
        transform=ax.get_xaxis_transform(), fontsize=9, color="gray", alpha=0.6)
ax.text((TIER01 + TIER12) / 2, 0.975, "Tier 1", ha="center", va="bottom",
        transform=ax.get_xaxis_transform(), fontsize=9, color="gray", alpha=0.6)
ax.text((TIER12 + xmax) / 2,   0.975, "Tier 2", ha="center", va="bottom",
        transform=ax.get_xaxis_transform(), fontsize=9, color="gray", alpha=0.6)

ax.set_yticks(y)
ax.set_yticklabels(countries)
ax.set_xlabel("SDGI score")
ax.set_title(f"SDGI score: {YEAR_FROM} → {YEAR_TO}")

legend_handles = [mpatches.Patch(color=col, label=f"Tier {t}") for t, col in tier_colors.items()]
ax.legend(handles=legend_handles, title="Tier", loc="upper left")

highlight_countries = {"Greece", "Hungary", "Lithuania", "Malta", "Germany", "Norway"}
labels = []
for c in countries:
    if c in highlight_countries:
        labels.append(c)  
    else:
        labels.append(c)

ax.set_yticks(y)
ax.set_yticklabels(labels)

for ticklabel, c in zip(ax.get_yticklabels(), countries):
    if c in highlight_countries:
        ticklabel.set_fontweight("bold")

plt.tight_layout()
plt.show()
```
